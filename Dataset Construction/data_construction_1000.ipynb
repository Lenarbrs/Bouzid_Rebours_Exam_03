# 1. Library and data import



```python
import pandas as pd
```

```python
df_fake = pd.read_csv('Fake.csv') # Importing the dataset containing unreliable news
df_true = pd.read_csv('True.csv') # Importing the dataset containing truthful news
```

# 2. Data manipulation

```python
# Drop duplicate titles and texts in each dataset to ensure each article only appears once
df_fake = df_fake.drop_duplicates(subset='title')
df_fake = df_fake.drop_duplicates(subset='text')
df_true = df_true.drop_duplicates(subset='title')
df_true = df_true.drop_duplicates(subset='text')
print(len(df_fake), len(df_true))
```

```python
df_fake.loc[:, 'fake'] = 1 # Set a "fake" label to the fake news
df_true.loc[:, 'fake'] = 0 # Set a "true" label to the real news
```

```python
print((df_fake['fake'] == 1).all(), (df_true['fake'] == 0).all()) # Making sure all rows are correctly labelled
```

```python
df_fake_sample = df_fake.sample(n=500, random_state=42) # Randomly selecting 500 rows of the fake news dataset
df_true_sample = df_true.sample(n=500, random_state=42) # Randomly selecting 500 rows of the real news dataset
```

```python
display(df_fake_sample, df_true_sample)
```

```python
df = pd.concat([df_fake_sample, df_true_sample], ignore_index=True) # Concatenate the datasets
```

```python
df
```

```python
df_shuffled = df.sample(frac=1, random_state=42).reset_index(drop=True) # Shuffling the data to mix real and fake news and make the LLM annotation easier and safer in the case of an interrupt
```

```python
df_shuffled
```

```python
df_shuffled.to_csv('ISOT_1000_news.csv', index=False) # Saving the data
```
